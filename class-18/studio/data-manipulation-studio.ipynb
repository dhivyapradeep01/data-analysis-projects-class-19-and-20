## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!

import numpy as np
import pandas as pd

baltimore_df = pd.read_csv("dataset/baltimore_9-24-2016_9-30-2017.csv")
boston_df = pd.read_csv("dataset/boston_9-24-2016_9-30-2017.csv")
newyork_df = pd.read_csv("dataset/new-york_9-24-2016_9-30-2017.csv")
philly_df = pd.read_csv("dataset/philadelphia_9-24-2016_9-30-2017.csv")

# print(baltimore_df.head())
# print(boston_df.head())
# print(newyork_df.head())
# print(philly_df.head())

print('~~~~~~~Baltimore~~~~~~~')
print(baltimore_df.info())
print('~~~~~~~Boston~~~~~~~')
print(boston_df.info())
print('~~~~~~~New York~~~~~~~')
print(newyork_df.info())
print('~~~~~~~Philly~~~~~~~')
print(philly_df.info())
```

## Clean Your Data

In the last lesson, we cleaned the data related to San Francisco. Pull up your notebook from the last lesson and use it as a reference to clean up these new dataframes.

```python
# Clean your data here!

#set DATE to date type
baltimore_df['Date'] = pd.to_datetime(baltimore_df['Date'])
boston_df ['Date']= pd.to_datetime(boston_df['Date'])
newyork_df ['Date']= pd.to_datetime(newyork_df['Date'])
philly_df ['Date']= pd.to_datetime(philly_df['Date'])

#There are some empty columns we can drop once we combine so we can check for null in all dataframes together
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.


northeast_df = pd.concat([baltimore_df, philly_df], ignore_index=True)

```python
# Combine the four dataframes into one!
northeast_df = pd.concat([baltimore_df,boston_df,newyork_df, philly_df], ignore_index=True)
# Sort the combined dataframe by date
northeast_df = northeast_df.sort_values('Date')

# Reset the index
northeast_df = northeast_df.reset_index(drop=True)

print('Before cleanup')
print(northeast_df.info())

# Now clean it and drop all empty columns 
# Drop columns with all null values
northeast_df = northeast_df.dropna(axis=1, how='all')

print('After cleanup')
print(northeast_df.info())
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of **unit of sale** in the Northeast region? 
2. For each region, what is the average number of pumpkins per variety that came into terminal markets for the year? 

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
# Assuming northeast_df is already created and contains both datasets

q1_result = northeast_df.groupby('Unit of Sale').agg({
    'Low Price': 'mean',
    'High Price': 'mean'
}).reset_index()

print("Mean low and high prices for each type of unit of sale")
print(q1_result)
print("\n")

```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
q2_result = northeast_df.groupby(['City Name', 'Variety']).size().reset_index(name='Count')

# Now, calculate the average count per variety for each city
q2_average = q2_result.groupby('City Name').agg({
    'Count': 'mean'
}).reset_index()

print("Average number of pumpkins per variety for each city")
print(q2_average)
```
